```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

```

```python

# Load your dataset
df = pd.read_csv('canada_per_capita_income.csv')

```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])

# Predict income for a specific year (e.g., 2022)
year_to_predict = 2007
predicted_income = reg.predict([[year_to_predict]])
print(f'Predicted income for {year_to_predict}: ${predicted_income[0]:,.2f}')
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
import matplotlib.pyplot as plt

    # Plot the data points
    plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='*', label='Data Points')
    
    # Plot the linear regression line
    plt.plot(df['year'], reg.predict(df[['year']]), color='blue', linewidth=2, label='Linear Regression')
    
    # Customize the plot
    plt.xlabel('year')
    plt.ylabel('per capita income (US$)')
    plt.title('Linear Regression: Year vs. Per Capita Income')
    plt.legend()
    
    # Show the plot
    plt.show()
```
